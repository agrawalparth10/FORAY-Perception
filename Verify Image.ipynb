```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('image_4.txt',mode='r')
```

```python
text = []
for line in file:
    text.append(line)
```

```python
print(text)
```

```python
def convert(text):
    image = []
    for i in range(len(text)):
        new_text = text[i].replace(" ",",")
        next_text = new_text.split(",")
        num = []
        for n in next_text:
            if n == '' or n== '\n':
                continue
            num.append(int(n))
        image.append(num)
    return np.array(image)
```

```python
image_arr = convert(text)
```

```python
print(image_arr.shape)
```

```python
plt.imshow(image_arr,cmap = 'gray')
```
